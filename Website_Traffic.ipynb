```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("traffic.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['month'] = df['date'].dt.month
df['weekday'] = df['date'].dt.day_name()
```

```python
df.columns
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='event', data=df)
plt.title("Event Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(y='country', data=df)
plt.title("Top Countries")
plt.show()
```

```python
top_tracks = df['track'].value_counts().head(10)

sns.barplot(x=top_tracks.values, y=top_tracks.index)
plt.title("Top 10 Tracks")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

month_counts = df['month'].value_counts().sort_index()

plt.figure(figsize=(8,5))
sns.barplot(x=month_counts.index, y=month_counts.values)
plt.title("Website Traffic by Month")
plt.xlabel("Month")
plt.ylabel("Number of Visits")
plt.show()
```

```python
weekday_counts = df['weekday'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=weekday_counts.index, y=weekday_counts.values)
plt.title("Traffic by Weekday")
plt.xlabel("Day of Week")
plt.ylabel("Number of Visits")
plt.xticks(rotation=45)
plt.show()
```

```python
top_countries = df['country'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title("Top 10 Countries by Traffic")
plt.xlabel("Visits")
plt.ylabel("Country")
plt.show()
```

```python
top_cities = df['city'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top_cities.values, y=top_cities.index)
plt.title("Top Cities by Traffic")
plt.xlabel("Visits")
plt.ylabel("City")
plt.show()
```

```python
top_artists = df['artist'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top_artists.values, y=top_artists.index)
plt.title("Top Artists by Traffic")
plt.xlabel("Number of Clicks")
plt.ylabel("Artist")
plt.show()
```

```python
top_tracks = df['track'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=top_tracks.values, y=top_tracks.index)
plt.title("Top Tracks")
plt.xlabel("Clicks")
plt.ylabel("Track")
plt.show()
```

```python
event_country = pd.crosstab(df['event'], df['country'])

event_country.plot(kind='bar', figsize=(10,6))
plt.title("Event Distribution by Country")
plt.show()
```
